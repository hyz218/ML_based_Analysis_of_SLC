# Figure 4. OncoPrinter

```python
import pandas as pd
```

## Load Data

```python
df = pd.read_csv('New_Data/TCGA_GTEX_SLC_103.csv')
df_001 = pd.read_csv('New_Data/SLC_biomarker_innerjoin_001.csv')
```

## Preprocessing

```python
df_slc = df_001.loc[:, ['SLC','category', 'expression']]
# column 생성
df_slc['expression'] = 'HIGH'
df_slc['EXP'] = 'EXP'
# value_counts를 column으로 생성
df_slc['count'] = df_slc.groupby(['category'])['SLC'].transform('count')
# 정렬
df_slc.sort_values(by='count', ascending=False, inplace=True)
# 이름 뒤에 cancer 붙이기
df_slc['category'] = df_slc['category'] + '_cancer'
# type = CNA로 수정
df_slc['EXP'] = 'CNA'
# expression = AMP: 빨간색 / GAIN: 분홍색
df_slc.loc[df_slc['count']>10, 'expression'] = 'AMP'
df_slc.loc[df_slc['count']<=10, 'expression'] = 'GAIN'
# drop column
df_slc.drop(columns=['count'], inplace=True)
```

```python
# 파일 내보내기
df_slc.to_csv('New_Data/oncoprinter.csv', index=False, header=None)
```
