## Load Data

```python
# 파일 불러오기 (biomarker 기준에 따라)
df = pd.read_csv('./New_Data/TCGA_GTEX_SLC_103.csv')
# df = pd.read_csv('./New_Data/TCGA_GTEX_SLC_.csv')
```

```python
# 2. 19개 암종 + GTEX + SLC bio-marker
df_biomarker = df.drop(columns=['sample','TCGA_GTEX_main_category','cancer','label'])
df_biomarker.rename(columns={'label_GTEx_100':'label'}, inplace=True)
```

### SVM

```python
df = df_biomarker

X = df.iloc[:,:-1]
y = df.iloc[:,-1:]
label_num = len(y['label'].unique())

# make train test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=123)

svm_model = svm.SVC(kernel='linear')
svm_model.fit(X_train, y_train)
y_pred_SVM = svm_model.predict(X_test)

zipped = zip(svm_model.coef_[0], df.columns[:-1])
zipped = sorted(zipped, key = lambda t: t[1])
imp_svm = pd.DataFrame(zipped).sort_values(0, ascending=False).reset_index(drop=True)
```

```python
pd.Series(imp_svm[0].tolist(), index=imp_svm[1].tolist()).plot(kind='barh', figsize = (20, 100))
plt.title('SVM Feature Importance')
plt.savefig('Figure/svm_feature_importance.png')
```

### XGBoost
